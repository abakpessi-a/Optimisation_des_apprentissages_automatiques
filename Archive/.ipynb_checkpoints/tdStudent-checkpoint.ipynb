```python
#Suppression de la colonne student_id
df = df.drop(columns=["student_id"])
```

```python
#Séparation des donnnées
y = df["exam_score"]  #On stock la colonne à prédire 'exam_score' dans y
X = df.drop(columns=["exam_score"]) #On stock les colonnes autres que celle à prédire 'features' dans X
```

```python
# Encoder toutes les colonnes object/category
X = pd.get_dummies(X, drop_first=False) #get_dummies()remplace toutes les colonnes catégorielles de X par des colonnes binaires
```

```python
#Vérification de l'encodage
X.head()
```

```python
# Conversion en listes Python
X_list = X.values.tolist()
y_true = y.values.tolist()

# Contrôles: on vérifie que X_list et y_true ont même nombre de valeurs
m = len(X_list)
n_features = len(X_list[0])
assert m == len(y_true), "Incohérence tailles X / y"
print(f"m = {m} exemples | n_features = {n_features} variables")
```

```python
#Implétons la fonction de prédiction
def prediction(w0, w, x):
    """
    y_hat = w0 + sum_j (w_j * x_j)
    - w0 : biais (float)
    - w  : liste[float] de taille n
    - x  : liste[float] de taille n
    """
    #On controle la longueur w et x
    if len(w) != len(x):
        raise ValueError(f"len(w)={len(w)} != len(x)={len(x)}")
    return w0 + sum(w_j * x_j for w_j, x_j in zip(w, x))
#Implémenton le MSE
def mse(y_true, y_pred):
    """
    J = (1/m) * sum_i (y_i - yhat_i)^2
    """
    y_true = list(y_true)
    y_pred = list(y_pred)
    #On vérifie que la liste y_true n'est pas vide
    if len(y_true) == 0:
        raise ValueError("y_true est vide (m = 0).")
    #y_true et y_pred doivent avoir même nombre de valeurs    
    if len(y_true) != len(y_pred):
        raise ValueError(f"Tailles différentes : len(y_true)={len(y_true)} != len(y_pred)={len(y_pred)}")
    m = len(y_true)
    se = sum((yt - yp) ** 2 for yt, yp in zip(y_true, y_pred))
    return se / m
```

```python
#Calcule des dérivées partielles
def compute_gradients(w0, w, X, y_true, predict_fn):
    """
    Retourne : grad_w0 (float), grad_w (liste[float] de taille n)
    """
    m = len(X)
    # 1) prédictions et résidus e_i = y_i - yhat_i
    y_pred = [predict_fn(w0, w, x_i) for x_i in X]
    residuals = [yt - yp for yt, yp in zip(y_true, y_pred)]

    # 2) gradient du biais
    grad_w0 = -(2 / m) * sum(residuals)

    # 3) gradient des coefficients
    n = len(w)
    grad_w = []
    for j in range(n):
        s = 0.0
        for i in range(m):
            s += X[i][j] * residuals[i]
        grad_w.append(-(2 / m) * s)

    return grad_w0, grad_w
```
